```python
%%cython
#solution to problem 1
cimport numpy as np
import numpy as np
from libc.math cimport fabs, exp
cimport cython

cdef double g(double x, double sigma):
    #The absolute value isn't necessary here
    return exp(-(x/sigma)**2)

@cython.boundscheck(False)
@cython.wraparound(False)
def perona_malik3(np.ndarray[dtype=double,ndim=2] u, double lbda, double sigma, int reps):
    cdef int n,i,j, height=u.shape[0], width=u.shape[1]
    cdef double temp, dif1, dif2, dif3, dif4
    cdef np.ndarray[dtype=double,ndim=1] temprow = np.empty(width)
    for n in xrange(reps):
        temprow[:]=u[0] #python slicing, possibly slow. Maybe do it with a memoryview??
        for i in xrange(1,height-1):
            temp=u[i,0]
            for j in xrange(1,width-1):
                dif1=temprow[j]-u[i,j]
                dif2=u[i+1,j]-u[i,j]
                dif3=temprow[j-1]-u[i,j]
                dif4=u[i,j+1]-u[i,j]
                temprow[j],u[i,j]=u[i,j],u[i,j]+lbda*(g(dif1,sigma)*dif1+g(dif2,sigma)*dif2+g(dif3,sigma)*dif3+g(dif4,sigma)*dif4)
```

```python
%%cython
#solution to problem 2
cimport numpy as np
import numpy as np
from libc.math cimport fabs, exp
cimport cython

cdef inline double g(double x, double sinv):
    #in this case the absolute value is not necessary.
    return exp(-(x*sinv)**2)

@cython.boundscheck(False)
@cython.wraparound(False)
def perona_malik9(np.ndarray[dtype=double,ndim=2] u, double lbda, double sigma, int reps):
    cdef int n,i,j, height=u.shape[0], width=u.shape[1]
    cdef double dif1, dif2, dif3, dif4, sinv=1./sigma
    cdef np.ndarray[dtype=double,ndim=1] temprow = np.empty(width)
    for n in xrange(reps):
        #do the first row
        #do the first item in the first row
        dif1,dif2=u[1,0]-u[0,0],u[0,1]-u[0,0]
        temprow[0],u[0,0]=u[0,0],u[0,0]+lbda*(g(dif1,sinv)*dif1+g(dif2,sinv)*dif2)
        #do the middle entries of the first row
        for j in xrange(1,width-1):
            dif1,dif2,dif3=u[i+1,j]-u[i,j],temprow[j-1]-u[i,j],u[i,j+1]-u[i,j]
            temprow[j],u[i,j]=u[i,j],u[i,j]+lbda*(g(dif1,sinv)*dif1+g(dif2,sinv)*dif2+g(dif3,sinv)*dif3)
        #do the last entry in the first row
        dif1,dif2=u[1,width-1]-u[0,width-1],temprow[width-2]-u[0,width-1]
        temprow[width-1],u[0,width-1]=u[0,width-1],u[0,width-1]+lbda*(g(dif1,sinv)*dif1+g(dif2,sinv)*dif2)
        #do the middle rows
        for i in xrange(1,height-1):
            #do the first entry of the current row
            dif1,dif2,dif3=temprow[0]-u[i,0],u[i+1,0]-u[i,0],u[i,1]-u[i,0]
            temprow[0],u[i,0]=u[i,0],u[i,0]+lbda*(g(dif1,sinv)*dif1+g(dif2,sinv)*dif2+g(dif3,sinv)*dif3)
            #do the middle entries of the current row
            for j in xrange(1,width-1):
                dif1,dif2,dif3,dif4=temprow[j]-u[i,j],u[i+1,j]-u[i,j],temprow[j-1]-u[i,j],u[i,j+1]-u[i,j]
                temprow[j],u[i,j]=u[i,j],u[i,j]+lbda*(g(dif1,sinv)*dif1+g(dif2,sinv)*dif2+g(dif3,sinv)*dif3+g(dif4,sinv)*dif4)
            #do the last entry of the current row
            dif1,dif2,dif3=temprow[width-1]-u[i,width-1],u[i+1,width-1]-u[i,width-1],temprow[width-2]-u[i,width-1]
            temprow[width-1],u[i,width-1]=u[i,width-1],u[i,width-1]+lbda*(g(dif1,sinv)*dif1+g(dif2,sinv)*dif2+g(dif3,sinv)*dif3)
        #do the first entry of the last row
        dif1,dif2=temprow[0]-u[height-1,0],u[height-1,1]-u[height-1,0]
        temprow[0],u[height-1,0]=u[height-1,0],u[height-1,0]+lbda*(g(dif1,sinv)*dif1+g(dif2,sinv)*dif2)
        #do the middle entries of the last row
        for j in xrange(1,width-1):
            dif1,dif2,dif3=temprow[j]-u[height-1,j],temprow[j-1]-u[height-1,j],u[height-1,j+1]-u[height-1,j]
            temprow[j],u[height-1,j]=u[height-1,j],u[height-1,j]+lbda*(g(dif1,sinv)*dif1+g(dif2,sinv)*dif2+g(dif3,sinv)*dif3)
        #do the last entry of the last row
        dif1,dif2=temprow[width-1]-u[height-1,width-1],temprow[width-2]-u[height-1,width-1]
        #we don't actually need to store the last temp value.
        u[height-1,width-1]+=lbda*(g(dif1,sinv)*dif1+g(dif2,sinv)*dif2)
```

```python
%%cython
#solution to problem 3
cimport numpy as np
import numpy as np
from libc.math cimport fabs, exp, fmax
cimport cython

cdef inline double g(double x, double sinv):
    #absolute value isn't necessary here.
    return exp(-(x*sinv)**2)

@cython.boundscheck(False)
@cython.wraparound(False)
cdef inline double norm(double[:] x, double[:] y):
    #using the infinity norm
    return fmax(fmax(x[0]-y[0],x[1]-y[1]),x[2]-y[2])

@cython.boundscheck(False)
@cython.wraparound(False)
def perona_malik_color2(np.ndarray[dtype=double,ndim=3] U, double lbda, double sigma, int reps):
    #this one isn't very readable, but works the same as the black and white version except
    #that instead of a single value for a pixel, now we have three to deal with 3 dimensional vectors
    #we use the norm function defined above for the computation of g at each point.
    cdef int n,i,j,k, height=U.shape[0], width=U.shape[1]
    cdef double g1, g2, g3, g4, cst=4./3., sinv=1./sigma, temp
    cdef double[:,:,:] u = U
    cdef double[:,:] temprow = np.empty((width,3))
    for n in xrange(reps):
        g1=g(norm(u[1,0],u[0,0]),sinv)
        g2=g(norm(u[0,1],u[0,0]),sinv)
        for k in xrange(3):
            temprow[0,k],u[0,0,k]=u[0,0,k],u[0,0,k]+lbda*(g1*(u[1,0,k]-u[0,0,k])+g2*(u[0,1,k]-u[0,0,k]))
        for j in xrange(1,width-1):
            g1=g(norm(u[i+1,j],u[i,j]),sinv)
            g2=g(norm(temprow[j-1],u[i,j]),sinv)
            g3=g(norm(u[i,j+1],u[i,j]),sinv)
            for k in xrange(3):
                temprow[j,k],u[i,j,k]=u[i,j,k],u[i,j,k]+lbda*(g1*(u[i+1,j,k]-u[i,j,k])+g2*(temprow[j-1,k]-u[i,j,k])+g3*(u[i,j+1,k]-u[i,j,k]))
        g1=g(norm(u[1,width-1],u[0,width-1]),sinv)
        g2=g(norm(temprow[width-2],u[0,width-1]),sinv)
        for k in xrange(3):
            temprow[width-1,k],u[0,width-1,k]=u[0,width-1,k],u[0,width-1,k]+lbda*(g1*(u[1,width-1,k]-u[0,width-1,k])+g2*(temprow[width-2,k]-u[0,width-1,k]))
        for i in xrange(1,height-1):
            g1=g(norm(temprow[0],u[i,0]),sinv)
            g2=g(norm(u[i+1,0],u[i,0]),sinv)
            g3=g(norm(u[i,1],u[i,0]),sinv)
            for k in xrange(3):
                temprow[0,k],u[i,0,k]=u[i,0,k],u[i,0,k]+lbda*(g1*(temprow[0,k]-u[i,0,k])+g2*(u[i+1,0,k]-u[i,0,k])+g3*(u[i,1,k]-u[i,0,k]))
            for j in xrange(1,width-1):
                g1=g(norm(temprow[j],u[i,j]),sinv)
                g2=g(norm(u[i+1,j],u[i,j]),sinv)
                g3=g(norm(temprow[j-1],u[i,j]),sinv)
                g4=g(norm(u[i,j+1],u[i,j]),sinv)
                for k in xrange(3):
                    temprow[j,k],u[i,j,k]=u[i,j,k],u[i,j,k]+lbda*(g1*(temprow[j,k]-u[i,j,k])+g2*(u[i+1,j,k]-u[i,j,k])+g3*(temprow[j-1,k]-u[i,j,k])+g4*(u[i,j+1,k]-u[i,j,k]))
            g1=g(norm(temprow[width-1],u[i,width-1]),sinv)
            g2=g(norm(u[i+1,width-1],u[i,width-1]),sinv)
            g3=g(norm(temprow[width-2],u[i,width-1]),sinv)
            for k in xrange(3):
                temprow[width-1,k],u[i,width-1,k]=u[i,width-1,k],u[i,width-1,k]+lbda*(g1*(temprow[width-1,k]-u[i,width-1,k])+g2*(u[i+1,width-1,k]-u[i,width-1,k])+g3*(temprow[width-2,k]-u[i,width-1,k]))
        g1=g(norm(temprow[0],u[height-1,0]),sinv)
        g2=g(norm(u[height-1,1],u[height-1,0]),sinv)
        for k in xrange(3):
            temprow[0,k],u[height-1,0,k]=u[height-1,0,k],u[height-1,0,k]+lbda*(g1*(temprow[0,k]-u[height-1,0,k])+g2*(u[height-1,1,k]-u[height-1,0,k]))
        for j in xrange(1,width-1):
            g1=g(norm(temprow[j],u[height-1,j]),sinv)
            g2=g(norm(temprow[j-1],u[height-1,j]),sinv)
            g3=g(norm(u[height-1,j+1],u[height-1,j]),sinv)
            for k in xrange(3):
                temprow[j,k],u[height-1,j,k]=u[height-1,j,k],u[height-1,j,k]+lbda*(g1*(temprow[j,k]-u[height-1,j,k])+g2*(temprow[j-1,k]-u[height-1,j,k])+g3*(u[height-1,j+1,k]-u[height-1,j,k]))
        g1=g(norm(temprow[width-1],u[height-1,width-1]),sinv)
        g2=g(norm(temprow[width-2],u[height-1,width-1]),sinv)
        for k in xrange(3):
            #we don't actually need to store the last difference
            u[height-1,width-1,k]+=lbda*(g1*(temprow[width-1,k]-u[height-1,width-1,k])+g2*(temprow[width-2,k]-u[height-1,width-1,k]))
```

```python
%%cython
#this is a function needed in my solution of problem 3
import numpy as np
cimport numpy as np
cimport cython
from libc.math cimport fmin, fmax, sqrt, fabs

float32=np.float32
ctypedef np.float32_t float32_T

@cython.boundscheck(False)
@cython.wraparound(False)
cdef inline bimin(float32_T[:] arr, int size):
    #changes the first two items of an array to be the
    #two smallest items in the array
    cdef int i
    cdef float32_T min1,min2
    if arr[0]<arr[1]:
        min1,min2=arr[0],arr[1]
    else:
        min2,min1=arr[0],arr[1]
    for i in xrange(2,size):
        if arr[i]<min1:
            min1,min2=arr[i],min1
        elif arr[1]<min2:
            min2=arr[i]
    arr[0]=min1
    arr[1]=min2

@cython.boundscheck(False)
@cython.wraparound(False)
cpdef makecoefs(np.ndarray[dtype=float32_T,ndim=2] vdifs,\
np.ndarray[dtype=float32_T,ndim=2] hdifs, np.ndarray[dtype=float32_T,ndim=2] ldifs,\
np.ndarray[dtype=float32_T,ndim=2] udifs, np.ndarray[dtype=float32_T,ndim=2] coefs):
    #This one isn't particularly well optimized
    #it *shouldn't* be too terrible though.
    #the operations are in place to fill the coeffs array
    #I don't know of any nice way of doing this, so heres a tractable way.
    cdef int i,j,k,h=coefs.shape[0],w=coefs.shape[1]
    cdef float32_T[:] temp=np.empty(8,dtype=np.float32)
    #do first point
    temp[0]=vdifs[0,0]
    temp[1]=ldifs[0,0]
    temp[2]=hdifs[0,0]
    bimin(temp,3)
    coefs[0,0]=(temp[0]+temp[1])*.5
    for j in xrange(1,w-1):
        temp[0]=hdifs[0,j-1]
        temp[1]=hdifs[0,j]
        temp[2]=udifs[0,j-1]
        temp[3]=vdifs[0,j]
        temp[4]=ldifs[0,j]
        bimin(temp,5)
        coefs[0,j]=(temp[0]+temp[1])*.5
    temp[0]=hdifs[0,w-2]
    temp[1]=udifs[0,w-2]
    temp[2]=vdifs[0,w-1]
    bimin(temp,3)
    coefs[0,0]=(temp[0]+temp[1])*.5
    for i in xrange(1,h-1):
        temp[0]=vdifs[i-1,0]
        temp[1]=vdifs[i,0]
        temp[2]=udifs[i-1,0]
        temp[3]=hdifs[i,0]
        temp[4]=ldifs[i,0]
        bimin(temp,5)
        coefs[i,0]=(temp[0]+temp[1])*.5
        for j in xrange(1,w-1):
            temp[0]=ldifs[i-1,j-1]
            temp[1]=ldifs[i,j]
            temp[2]=vdifs[i-1,j]
            temp[3]=vdifs[i,j]
            temp[4]=udifs[i-1,j]
            temp[5]=udifs[i,j-1]
            temp[6]=hdifs[i,j-1]
            temp[7]=hdifs[i,j]
            bimin(temp,8)
            coefs[i,j]=(temp[0]+temp[1])*.5
        temp[0]=vdifs[i-1,w-1]
        temp[1]=vdifs[i,w-1]
        temp[2]=ldifs[i-1,w-2]
        temp[3]=hdifs[i,w-2]
        temp[4]=udifs[i,w-2]
        bimin(temp,6)
        coefs[i,w-1]=(temp[0]+temp[1])*.5
    temp[0]=vdifs[h-2,0]
    temp[1]=udifs[h-2,0]
    temp[2]=hdifs[h-1,0]
    bimin(temp,3)
    coefs[h-1,0]=(temp[0]+temp[1])*.5
    for j in xrange(1,w-1):
        temp[0]=hdifs[h-1,j-1]
        temp[1]=hdifs[h-1,j]
        temp[2]=ldifs[h-2,j-1]
        temp[3]=vdifs[h-2,j]
        temp[4]=udifs[h-2,j]
        bimin(temp,5)
        coefs[h-1,j]=(temp[0]+temp[1])*.5
    temp[0]=hdifs[h-1,w-2]
    temp[1]=ldifs[h-2,w-2]
    temp[2]=vdifs[h-2,w-1]
    bimin(temp,3)
    coefs[h-1,w-1] = (temp[0]+temp[1])*.5
```

```python
#vanilla numpy version for problem 3
def min_bias_color2(u, lbda, reps):
    #this uses a good bit of excess memory...
    #the approach is based on the approach I took with the
    #optimized versions of problems 2 and 3 below
    difs=np.empty_like(u)
    uN=u[:-1]
    uS=u[1:]
    uW=u[:,:-1]
    uE=u[:,1:]
    uNW=u[:-1,:-1]
    uNE=u[:-1,1:]
    uSW=u[1:,:-1]
    uSE=u[1:,1:]
    Ndifs=difs[:-1]
    Sdifs=difs[1:]
    Wdifs=difs[:,:-1]
    Edifs=difs[:,1:]
    temp=np.empty_like(u)
    coefs=np.empty_like(temp[:,:,0])
    coefsN=coefs[:-1]
    coefsS=coefs[1:]
    coefsW=coefs[:,:-1]
    coefsE=coefs[:,1:]
    vtemp=temp[:-1]
    htemp=temp[:,:-1]
    dtemp=temp[:-1,:-1]
    vdifs=np.empty_like(vtemp[:,:,0])
    hdifs=np.empty_like(htemp[:,:,0])
    ldifs=np.empty_like(dtemp[:,:,0])
    udifs=np.empty_like(dtemp[:,:,0])
    lastrow=difs[-1]
    difsT=difs.T
    coefsT=coefs.T
    for i in xrange(reps):
        vtemp[:]=uN=uS
        np.fabs(vtemp,out=vtemp)
        np.max(vtemp,axis=2,out=vdifs)
        htemp[:]=uW-uE
        np.fabs(htemp,out=htemp)
        np.max(htemp,axis=2,out=hdifs)
        dtemp[:]=uNW-uSE
        np.fabs(dtemp,out=dtemp)
        np.max(dtemp,axis=2,out=ldifs)
        dtemp[:]=uSW-uNE
        np.fabs(dtemp,out=dtemp)
        np.max(dtemp,axis=2,out=udifs)
        makecoefs(vdifs,hdifs,ldifs,udifs,coefs)
        coefs*=lbda
        #overwrite most of difs instead of reinitializing
        Ndifs[:]=uS-uN
        #initialize what's left
        lastrow[:]=0.
        Sdifs+=uN
        Sdifs-=uS
        Wdifs+=uE
        Wdifs-=uW
        Edifs+=uW
        Edifs-=uE
        difsT*=coefsT#broadcasting
        u+=difs
```

```python
#numexpr optimized version
import numexpr as ne
def min_bias_color(u, lbda, reps):
    #this uses a good bit of excess memory...
    #the approach is based on the approach I took with the
    #optimized versions of problems 2 and 3 below
    difs=np.empty_like(u)
    lb32=np.float32(lbda)
    uN=u[:-1]
    uS=u[1:]
    uW=u[:,:-1]
    uE=u[:,1:]
    uNW=u[:-1,:-1]
    uNE=u[:-1,1:]
    uSW=u[1:,:-1]
    uSE=u[1:,1:]
    Ndifs=difs[:-1]
    Sdifs=difs[1:]
    Wdifs=difs[:,:-1]
    Edifs=difs[:,1:]
    temp=np.empty_like(u)
    coefs=np.empty_like(temp[:,:,0])
    coefsN=coefs[:-1]
    coefsS=coefs[1:]
    coefsW=coefs[:,:-1]
    coefsE=coefs[:,1:]
    vtemp=temp[:-1]
    htemp=temp[:,:-1]
    dtemp=temp[:-1,:-1]
    vdifs=np.empty_like(vtemp[:,:,0])
    hdifs=np.empty_like(htemp[:,:,0])
    ldifs=np.empty_like(dtemp[:,:,0])
    udifs=np.empty_like(dtemp[:,:,0])
    lastrow=difs[-1]
    difsT=difs.T
    coefsT=coefs.T
    for i in xrange(reps):
        ne.evaluate('abs(uN-uS)',out=vtemp)
        np.max(vtemp,axis=2,out=vdifs)
        ne.evaluate('abs(uW-uE)',out=htemp)
        np.max(htemp,axis=2,out=hdifs)
        ne.evaluate('abs(uNW-uSE)',out=dtemp)
        np.max(dtemp,axis=2,out=ldifs)
        ne.evaluate('abs(uSW-uNE)',out=dtemp)
        np.max(dtemp,axis=2,out=udifs)
        makecoefs(vdifs,hdifs,ldifs,udifs,coefs)
        ne.evaluate('lb32*coefs',out=coefs)
        #overwrite most of difs instead of reinitializing
        ne.evaluate('uS-uN',out=Ndifs)
        #initialize what's left
        lastrow[:]=0.
        ne.evaluate('Sdifs+uN-uS',out=Sdifs)
        ne.evaluate('Wdifs+uE-uW',out=Wdifs)
        ne.evaluate('Edifs+uW-uE',out=Edifs)
        difsT*=coefsT#broadcasting
        difs*=lb32
        ne.evaluate('u+difs',out=u)
```

```python
#here are some alternate solutions
#I figured these out after the fact.
#What I have in Cython above is probably the more obvious way to do it,
#that being said, these work well too.
#They are maybe a little less intuitive, but 
#usually faster than the normal iterative versions
#in all cases, these use more memory than the other ones
#these all expect float32 values, though the vanilla numpy
#versions should work with any sort of float or complex number.
#these are the best optimized versions I have found thus far.
#the idea is that instead of manually iterating over the array,
#we calculate differences between different views of the same data.
#this avoids excess computation and faster iteration time, but
#it also requires much more memory.
import numexpr as ne
import theano as ta
from theano import tensor as te
#currently, importing Theano breaks my cythonmagic function,
#so evaluate the previous cells first.
```

```python
#vanilla numpy black and white
#this one does not have any package dependencies besides numpy
def perona_malik11(u, lbda, sigma, reps):
    unew=u.copy()
    #make some views of the same data as u
    uN=u[:-1]
    uS=u[1:]
    uW=u[:,:-1]
    uE=u[:,1:]
    #make similar views for unew
    uNnew=unew[:-1]
    uSnew=unew[1:]
    uWnew=unew[:,:-1]
    uEnew=unew[:,1:]
    for i in xrange(reps):
        temp=lbda*np.exp((np.max(uS-uN,axis=2)/sigma)**2)*(uS-uN)
        uNnew+=temp
        uSnew-=temp
        temp=lbda*np.exp(((uE-uW)/sigma)**2)*(uE-uW)
        uWnew+=temp
        uEnew-=temp
        u[:]=unew
```

```python
#numexpr-ized black and white
#this one is blazing fast
def perona_malik12(u, lbda, sigma, reps):
    #expects array of 32 bit floats
    sinv=np.float32(1./sigma)
    lbda32=np.float32(lbda)
    unew=u.copy()
    #make some views of the same data as u
    uN=u[:-1]
    uS=u[1:]
    uW=u[:,:-1]
    uE=u[:,1:]
    #make similar views for unew
    uNnew=unew[:-1]
    uSnew=unew[1:]
    uWnew=unew[:,:-1]
    uEnew=unew[:,1:]
    for i in xrange(reps):
        temp=ne.evaluate('lbda32*exp((sinv*(uS-uN))**2)*(uS-uN)')
        ne.evaluate('uNnew+temp',out=uNnew)
        ne.evaluate('uSnew-temp',out=uSnew)
        temp=ne.evaluate('lbda32*exp((sinv*(uE-uW))**2)*(uE-uW)')
        ne.evaluate('uWnew+temp',out=uWnew)
        ne.evaluate('uEnew-temp',out=uEnew)
        u[:]=unew
```

```python
#black and white version using theano
#on my machine this is slower than the numexpr version
#on a machine where Theano can interface with the graphics card
#it could be faster.
#we first predefine the operator we need.
a=te.fmatrix('a')
b=te.fmatrix('b')
difference=a-b
dif=ta.function([a,b],difference)
lbda_scl=te.fscalar('lbda_s')
sinv_scl=te.fscalar('sigma_s')
operator14=lbda_scl*te.exp(((a-b)*sinv_scl)**2)*(a-b)
op14=ta.function([a,b,lbda_scl,sinv_scl],operator14)
def perona_malik14(u, lbda, sigma, reps):
    unew=u.copy()
    sinv=np.float32(sigma)
    lb32=np.float32(lbda)
    #make some views of the same data as u
    uN=u[:-1]
    uS=u[1:]
    uW=u[:-1]
    uE=u[1:]
    #make similar views for unew
    uNnew=unew[:-1]
    uSnew=unew[1:]
    uWnew=unew[:-1]
    uEnew=unew[1:]
    for i in xrange(reps):
        temp=op14(uS,uN,lb32,sinv)
        uNnew+=temp
        uSnew-=temp
        temp=op14(uE,uW,lb32,sinv)
        uWnew+=temp
        uEnew-=temp
        u[:]=unew
```

```python
#vanilla numpy color implementation
def perona_malik_color20(u,lbda,sigma,iters):
    unew=u.copy()
    #transposes are necessary so that broadcasting can work later.
    uNT=u[:-1].T
    uST=u[1:].T
    uWT=u[:,:-1].T
    uET=u[:,1:].T
    uNnewT=unew[:-1].T
    uSnewT=unew[1:].T
    uWnewT=unew[:,:-1].T
    uEnewT=unew[:,1:].T
    vdifs=np.empty_like(uNT[0])
    hdifs=np.empty_like(uWT[0])
    temp=np.empty_like(u)
    vtemp=temp[:-1].T
    htemp=temp[:,:-1].T
    temp2=np.empty_like(temp)
    vtemp2=temp2[:-1].T
    htemp2=temp2[:,:-1].T
    for n in xrange(iters):
        #I lined the computations up in this way so as to 
        #avoid unnecessary memory allocation and deallocation
        #as we go through the for loop.
        vtemp[:]=uST-uNT
        np.fabs(vtemp,out=vtemp2)
        np.max(vtemp2,axis=0,out=vdifs)
        vdifs/=sigma
        np.square(vdifs,out=vdifs)
        vdifs*=-1
        np.exp(vdifs,out=vdifs)
        vdifs*=lbda
        vtemp*=vdifs#broadcasting across color dimension
        uNnewT+=vtemp
        uSnewT-=vtemp
        htemp[:]=uET-uWT
        np.fabs(htemp,out=htemp2)
        np.max(htemp2,axis=0,out=hdifs)
        hdifs/=sigma
        np.square(hdifs,out=hdifs)
        hdifs*=-1
        np.exp(hdifs,out=hdifs)
        hdifs*=lbda
        htemp*=hdifs#broadcasting across color dimension
        uWnewT+=htemp
        uEnewT-=htemp
        u[:]=unew
```

```python
#numexpr optimized color version
import numexpr as ne
def perona_malik_color13(u,lbda,sigma,iters):
    sinv=np.float32(1./sigma)
    lbda32=np.float32(lbda)
    unew=u.copy()
    uNT=u[:-1].T
    uST=u[1:].T
    uWT=u[:,:-1].T
    uET=u[:,1:].T
    uNnewT=unew[:-1].T
    uSnewT=unew[1:].T
    uWnewT=unew[:,:-1].T
    uEnewT=unew[:,1:].T
    vdifs=np.empty_like(uNT[0])
    hdifs=np.empty_like(uWT[0])
    temp=np.empty_like(u)
    vtemp=temp[:-1].T
    htemp=temp[:,:-1].T
    temp2=np.empty_like(temp)
    vtemp2=temp2[:-1].T
    htemp2=temp2[:,:-1].T
    for n in xrange(iters):
        ne.evaluate('uST-uNT',out=vtemp)
        ne.evaluate('abs(vtemp)',out=vtemp2)
        np.max(vtemp2,axis=0,out=vdifs)
        ne.evaluate('lbda32*exp(-(sinv*(vdifs))**2)',out=vdifs)
        vtemp*=vdifs
        ne.evaluate('uNnewT+vtemp',out=uNnewT)
        ne.evaluate('uSnewT-vtemp',out=uSnewT)
        ne.evaluate('uET-uWT',out=htemp)
        ne.evaluate('abs(htemp)',out=htemp2)
        np.max(htemp2,axis=0,out=hdifs)
        ne.evaluate('lbda32*exp(-(sinv*(hdifs))**2)',out=hdifs)
        htemp*=hdifs
        ne.evaluate('uWnewT+htemp',out=uWnewT)
        ne.evaluate('uEnewT-htemp',out=uEnewT)
        u[:]=unew
```

```python
#theano color implementation
#here we predefine the operator we need
fmatrix3= te.TensorType('float32',(False,)*3)
a=fmatrix3('a')
b=fmatrix3('b')
lbda_scl=te.fscalar('lbda_scl')
sinv_scl=te.fscalar('sinv_scl')
operator16=(lbda_scl*te.exp((te.max(abs(a-b),axis=2)*sinv_scl)**2).T*(a-b).T).T
op16=ta.function([a,b,lbda_scl,sinv_scl],operator16)
def perona_malik_color16(u, lbda, sigma, reps):
    lb32=np.float32(lbda)
    sinv=np.float32(1./sigma)
    unew=u.copy()
    uN=u[:-1]
    uS=u[1:]
    uW=u[:,:-1]
    uE=u[:,1:]
    uNnew=unew[:-1]
    uSnew=unew[1:]
    uWnew=unew[:,:-1]
    uEnew=unew[:,1:]
    temp=np.empty_like(u)
    vtemp=temp[:-1].T
    htemp=temp[:,:-1].T
    for i in xrange(reps):
        vtemp=op16(uS,uN,lb32,sinv)
        uNnew+=vtemp
        uSnew-=vtemp
        htemp=op16(uE,uW,lb32,sinv)
        uWnew+=htemp
        uEnew-=htemp
        u[:]=unew
```
